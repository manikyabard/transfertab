```python
import pandas as pd
import matplotlib.pyplot as plt
import wandb
# import seaborn as sns
```

```python
api = wandb.Api()
run_1c_modelC = api.run("manikya/transfertab/6566qfds")
run_1c_modelB = api.run("manikya/transfertab/3ruoxpym")
```

```python
modelC_1c = run_1c_modelC.history()
```

```python
modelB_1c = run_1c_modelB.history()
```

```python
plt.rcParams.update({'font.size': 12})
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12
```

```python
plt.plot(modelB_1c['_step'], modelB_1c['train_loss'], label='Model B', color="#638c51", linewidth=3)
plt.plot(modelC_1c['_step'], modelC_1c['train_loss'], label='Model C', color='#fc870a', linewidth=3)
plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.savefig('../data/figures/exp1c.png', format='png', dpi=600)
```
